```python
# prompt: Genera un codigo que me permita graficar un vector en 3d conociendo sus componentes. Y que se pueda visualizar las componentes usadas. Tambine muestra el modulo del vector. Restringue el grafico al primer octante

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Definir las componentes del vector
x_comp = 2
y_comp = 3
z_comp = 4

# Crear la figura y el eje 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar el vector
ax.quiver(0, 0, 0, x_comp, y_comp, z_comp, color='blue', arrow_length_ratio=0.1)
# Graficar el segundo vector
ax.quiver(0, 0, 0, 2.1, 3.1, 4.1, color='yellow', arrow_length_ratio=0.1)

# Mostrar las componentes del vector
ax.text(x_comp, y_comp, z_comp, f'({x_comp}, {y_comp}, {z_comp})', color='red')

# Calcular y mostrar el mdulo del vector
modulo = np.sqrt(x_comp**2 + y_comp**2 + z_comp**2)
ax.text(x_comp / 2, y_comp / 2, z_comp / 2, f'Mdulo: {modulo:.2f}', color='green')

# Establecer lmites del grfico al primer octante
ax.set_xlim([0, max(x_comp, 1)])  # Ajusta el lmite mximo segn la componente x
ax.set_ylim([0, max(y_comp, 1)])  # Ajusta el lmite mximo segn la componente y
ax.set_zlim([0, max(z_comp, 1)])  # Ajusta el lmite mximo segn la componente z


# Etiquetas de los ejes
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

# Mostrar el grfico
plt.show()
```
