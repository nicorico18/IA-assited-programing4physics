**PRÁCTICA DE VECTORES EN EL ESPACIO**

```python
# prompt: genera una tabla de estos datos ordenados y un grafico donde se representen 3 vectores x,y y z y asignales el valor de 19,4 ;17,7;20,5. correspondientemente y el angulo de alfa, beta y gama es de 54,58,53 grados correspondientemente y el modulo debes calcularlo

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Datos de los vectores
data = {
    'Vector': ['x', 'y', 'z'],
    'Componente': [19.4, 17.7, 20.5],
    'Angulo (grados)': [54, 58, 53]
}

df = pd.DataFrame(data)
print(df)

# Cálculo del módulo (magnitud) del vector resultante
modulo = np.sqrt(19.4**2 + 17.7**2 + 20.5**2)

# Conversión de ángulos a radianes
alpha = np.radians(54)
beta = np.radians(58)
gamma = np.radians(53)


# Cálculo de las componentes cartesianas (usando ángulos como dirección)
x = modulo * np.sin(alpha) * np.cos(beta)
y = modulo * np.sin(alpha) * np.sin(beta)
z = modulo * np.cos(alpha)


# Crear la figura y los ejes 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


# Graficar el vector resultante
ax.quiver(0, 0, 0, x, y, z, color='r', label='Vector resultante')

# Graficar las componentes
ax.quiver(0, 0, 0, 19.4, 0, 0, color='b', label='Componente x')
ax.quiver(0, 0, 0, 0, 17.7, 0, color='g', label='Componente y')
ax.quiver(0, 0, 0, 0, 0, 20.5, color='m', label='Componente z')

# Ajustar los límites de los ejes
ax.set_xlim([0, max(19.4, x) + 5])
ax.set_ylim([0, max(17.7, y) + 5])
ax.set_zlim([0, max(20.5, z) + 5])

# Etiquetas de los ejes y leyenda
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()

# Mostrar el gráfico
plt.show()
```

Máquina de Atwood


Grafique y analice la función aceleracion vs inverso de la masa con minimos cuadrados. Descubra el significado de la pendiente y la relacion con la Primera ley de Newton.

```python
# prompt: grafica una función de aceleracion vs inverso de la masa con minimos cuadrados. Los datos son los siguientes: en el eje x irá el inverso de la masa que es de 4,85 kg y en el eje y irá el valor de la aceleración que es de 0,493m/s2 y como pendiente tengo el valor de 0,27kg. Como datos adicionales tengo la fuerza de 0,137N, distancia=0,5m y el tiempo promedio de=1,424s

import matplotlib.pyplot as plt
import numpy as np

# Datos proporcionados
inverso_masa = 1/4.85  # Inverso de la masa en kg^-1
aceleracion = 0.493   # Aceleración en m/s^2
pendiente = 0.27      # Pendiente de la recta de ajuste

# Crear los datos para la gráfica
x = np.array([inverso_masa])
y = np.array([aceleracion])


# Ajustar una línea a los datos usando mínimos cuadrados
# En este caso, como solo tenemos un punto, la línea pasa directamente por él.
# La pendiente es la proporcionada en el enunciado.
# El intercepto lo calculamos usando la fórmula de la recta: y = mx + b, por lo tanto b = y - mx
intercepto = aceleracion - pendiente * inverso_masa


# Crear la gráfica
plt.plot(x, y, 'ro', label='Datos Experimentales')  # Marcador para el dato experimental

# Crear la recta de mínimos cuadrados
x_recta = np.linspace(0, 0.3, 10)  # Rango de x para la recta
y_recta = pendiente * x_recta + intercepto
plt.plot(x_recta, y_recta, 'b-', label='Ajuste por Mínimos Cuadrados')

# Añadir etiquetas y título a la gráfica
plt.xlabel('1/Masa (kg⁻¹)')
plt.ylabel('Aceleración (m/s²)')
plt.title('Aceleración vs. Inverso de la Masa')
plt.legend()

# Mostrar la cuadrícula
plt.grid(True)

# Mostrar la gráfica
plt.show()


print(f"Pendiente de la recta de ajuste: {pendiente} N") #La pendiente representa la fuerza aplicada al sistema
print(f"Intercepto de la recta de ajuste: {intercepto} m/s^2")
```

#Ley de Hooke

```python
# prompt: Generar una tabla de pandas con Masa(kg):0.05,0.1,0.15,0.20,0.25,0.3,0.35 y Posicion(m):0.546,0.481,0.417,0.353,0.288,0.225,0.162

import pandas as pd

data = {
    'Masa(kg)': [0.05, 0.1, 0.15, 0.20, 0.25, 0.3, 0.35],
    'Posicion(m)': [0.546, 0.481, 0.417, 0.353, 0.288, 0.225, 0.162]
}

df = pd.DataFrame(data)
df
```

```python
# prompt: Calcula una columna Fuerza(N) que sea Masa(kg) multiplicado la acelaracion de la gravedad 9.8

df['Fuerza(N)'] = df['Masa(kg)'] * 9.8
df
```

```python
# prompt: Genera una columna Elongacion(m) que sea la Posicion(m) - 0.546
import numpy as np
df['Elongacion(m)'] = np.abs(df['Posicion(m)'] - 0.546)
df
```

```python
# prompt: Generar una columna F/dy(N/m) que sea la razon entre Fuerza(N) y Elongacion(m)

df['F/dy(N/m)'] = df['Fuerza(N)'] / df['Elongacion(m)']
df
```

```python
# prompt: Genera un grafico de Fuerza en funcion de la elongacion

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Datos proporcionados (asumiendo que ya tienes el DataFrame 'df' con las columnas calculadas)
data = {
    'Masa(kg)': [0.05, 0.1, 0.15, 0.20, 0.25, 0.3, 0.35],
    'Posicion(m)': [0.546, 0.481, 0.417, 0.353, 0.288, 0.225, 0.162]
}

df = pd.DataFrame(data)
df['Fuerza(N)'] = df['Masa(kg)'] * 9.8
df['Elongacion(m)'] = np.abs(df['Posicion(m)'] - 0.546)

# Crear la gráfica de Fuerza vs. Elongación
plt.plot(df['Elongacion(m)'], df['Fuerza(N)'], 'ro', label='Datos Experimentales')

# Ajuste lineal por mínimos cuadrados
pendiente, intercepto = np.polyfit(df['Elongacion(m)'], df['Fuerza(N)'], 1)
x_recta = np.linspace(0, max(df['Elongacion(m)']), 100)
y_recta = pendiente * x_recta + intercepto
plt.plot(x_recta, y_recta, 'b-', label=f'Ajuste (k = {abs(pendiente):.2f} N/m)')


# Agregar etiquetas y título a la gráfica
plt.xlabel('Elongación (m)')
plt.ylabel('Fuerza (N)')
plt.title('Fuerza vs. Elongación (Ley de Hooke)')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()

print(f"Constante del resorte (k): {abs(pendiente):.2f} N/m")
print(f"Intercepto: {intercepto:.2f} N")
```
