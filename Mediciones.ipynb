# Generar tabla de datos en colab

```python
 #prompt: Porfavor genera una tabla de datos para medir 5 veces las dimensiones de un prisma

import pandas as pd

# Crear un DataFrame vacío
data = {'Largo (mm)': [60.10, 60.45, 60.30, 60.10, 60.20],
        'Ancho (mm)': [62.35, 63.05, 62.30, 62.20, 62.25],
        'Alto (mm)': [39.15, 39.10, 39.15, 39.10, 39.15],
        "Profundidad (mm)": [24.15 , 24.35, 24.30, 24.40, 24.40],
        "Diametro (mm)": [19.30 , 19.25, 9.35, 19.45, 19.40]}
df = pd.DataFrame(data)

# Mostrar la tabla
df
```

```python
# prompt: Por favor calcula la desviacion estandar de cada columna

# Calcular la desviación estándar de cada columna
desviacion_estandar = df.std()

# Mostrar la desviación estándar
desviacion_estandar
```

```python
# Calcular la desviación estándar de cada columna
desviacion_estandar = df.std()

# Calcular el promedio de cada columna
promedio = df.mean()

# Crear un DataFrame con la desviación estándar y el promedio
resultados = pd.DataFrame({'Promedio': promedio, 'Desviación Estándar': desviacion_estandar})

# Mostrar la tabla de resultados
print("\nTabla de resultados:")
resultados
```

```python
# prompt: Calcula volumen del cubo con los promedios de LArgo ANcho y Alto

# Calcular el volumen del cubo usando los promedios de Largo, Ancho y Alto
volumen = promedio['Largo (mm)'] * promedio['Ancho (mm)'] * promedio['Alto (mm)']

# Imprimir el volumen
print(f"\nVolumen del cubo con los promedios de Largo, Ancho y Alto: {volumen} mm³")
```

```python
# prompt: Calcula el volumen del cilindro con Profundidad y diamtre promedio

import pandas as pd
import math

# ... (previous code remains the same)

# Calcular el volumen del cilindro usando los promedios de profundidad y diámetro
promedio_profundidad = promedio['Profundidad (mm)']
promedio_diametro = promedio['Diametro (mm)']
radio = promedio_diametro / 2
volumen_cilindro = math.pi * (radio**2) * promedio_profundidad

# Imprimir el volumen del cilindro
print(f"\nVolumen del cilindro con los promedios de Profundidad y Diámetro: {volumen_cilindro} mm³")
```

```python
# prompt: resta el volumen del cilindro del volumen del cubo y presentalo en metros cubicos

volumen_cubo = promedio['Largo (mm)'] * promedio['Ancho (mm)'] * promedio['Alto (mm)']
volumen_cilindro = math.pi * (promedio_diametro / 2)**2 * promedio_profundidad

diferencia_volumenes = volumen_cubo - volumen_cilindro

print(f"\nDiferencia de volumenes (cubo - cilindro): {diferencia_volumenes} mm³")

# Convertir la diferencia de volumenes a metros cúbicos
diferencia_volumenes_m3 = diferencia_volumenes / 1e9

print(f"\nDiferencia de volumenes (cubo - cilindro): {diferencia_volumenes_m3} m³")
```

```python
# prompt: obten la densidad si la masa es 0.127

masa = 0.127  # kg
volumen_cubo = promedio['Largo (mm)'] * promedio['Ancho (mm)'] * promedio['Alto (mm)']
volumen_cubo_m3 = volumen_cubo / 1e9

densidad = masa / volumen_cubo_m3
print(f"Densidad del cubo: {densidad} kg/m³")
```

```python
# prompt: calcula el error porcentual si la densidad tabulada es 940 kg/m^3

import pandas as pd
import math

# ... (previous code remains the same)

masa = 0.127  # kg
volumen_cubo = promedio['Largo (mm)'] * promedio['Ancho (mm)'] * promedio['Alto (mm)']
volumen_cubo_m3 = volumen_cubo / 1e9

densidad = masa / volumen_cubo_m3
densidad_tabulada = 940 # kg/m^3

error_porcentual = abs((densidad - densidad_tabulada) / densidad_tabulada) * 100

print(f"Densidad del cubo: {densidad} kg/m³")
print(f"Error porcentual: {error_porcentual:.2f}%")
```
